```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("Company_performance.csv")

X = data.iloc[:, [0]].values
y = data.iloc[:, 1].values

print(X, y)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X,y)
plt.plot(X, y, color = 'orange')
y_slr = lin_reg.predict(X)
plt.scatter(X, lin_reg.predict(X), color = 'blue')
plt.title("Degree 1 vs Degree 3")
plt.xlabel("No.of years")
plt.ylabel("No. of employees")

from sklearn.preprocessing import PolynomialFeatures

polf = PolynomialFeatures(degree = 2)
X_pol = polf.fit_transform(X)

plr = LinearRegression()
plr.fit(X_pol, y)
y_pred = plr.predict(X_pol)

plt.plot(X, y_pred, color = 'pink')

deg3 = PolynomialFeatures(degree = 3)
x3 = deg3.fit_transform(X)

p3 = LinearRegression()
p3.fit(x3, y)
y3 = p3.predict(x3)

plt.plot(X, y_pred, color = 'green')
plt.show()
```
